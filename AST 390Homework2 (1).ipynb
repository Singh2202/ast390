Stellar colors are expressed as the difference in astronomical magnitudes with the flux measured through different filters. The standard photometric system defines U, B, V, and more filters, where U is ultraviolet, B is blue, and V is visual, from which we can define a color index like $B-V$ (where here, $B$ is now the magnitude of the star through the B filter). $B-V$ is a difference in magnitudes, which can be expressed as

   $$B - V = -2.5 \log_{10}
  \left(\frac{f_B}{f_V}\right)$$

   where $f_B$ is the flux received from a star through the B filter and $f_V$
   is the flux from a star through the V filter.

   We can treat a star as a perfect blackbody (though this isn't *fully* true), and then
   take

   $$f_X \approx \int_0^\infty B_\lambda(T) s_X(\lambda) d\lambda$$

   where $B_\lambda(T)$ is the Planck function for temperature $T$, and $s_X$ is the filter X's response.
   The "$\approx$" here is because the distance to the star, its radius, and the fact that we need to
   integrate the Planck function over the outward angles all introduce a proportionality constant.  However, since
   we are taking the ratio of fluxes, this proportionality constant will cancel.
   
   We will take the filter to be a box-filter, with:

   $$s_X(\lambda) = \left \{ \begin{array}{cc} 1 & |\lambda - \lambda_{0,X}| \le  \frac{d\lambda_X}{2} \\
                                               0 & \mbox{otherwise} \end{array} \right . $$

   (i.e., the filter only lets light through in a narrow band of width $d\lambda_X$
   centered on a wavelength $\lambda_{0,X}$).  Reasonable values for the filter properties
   are:

   | filter   | $\lambda_{0,X}$  | $d\lambda_{X}$  |
   |----------|------------------|-----------------|
   |  B       |    445 nm        |     90 nm       |
   |  V       |    550 nm        |     90 nm       |
   

The function of the box filter in the integral for $f_X$ is essentially to "activate" the Planck function in the appropriate wavelength band, and to "kill" the Planck function everywhere else. Mathematically,
   
   $$f_X \approx \int_0^\infty B_\lambda(T) s_X(\lambda) d\lambda =  \int_{\lambda_{0,X} -\frac{d\lambda_X}{2}} ^{\lambda_{0,X} +\frac{d\lambda_X}{2}} B_\lambda(T) d\lambda$$
   
In particular, then,

$$f_B = \int_{400 \hspace{0.1em} \mathrm{nm}}^{490 \hspace{0.1em}\mathrm{nm}} B_\lambda(T) d\lambda \hspace{3em} \mathrm{and} \hspace{3em} f_V = \int_{505 \hspace{0.1em}\mathrm{nm}}^{595 \hspace{0.1em}\mathrm{nm}} B_\lambda(T) d\lambda$$

i.e. the filter acts simply to set the integration range (note: if our filter were instead Gaussian or Lorentzian or anything else, this wouldn't be true; this is a happy property of the box filter specificially). Now let's actually substitute the Planck function, $B_\lambda = \frac{2 h c^2}{\lambda^5} \frac{1}{\exp\left (\frac{hc}{\lambda k_B T}\right) - 1}$:

$$f_B = 2 h c^2 \int_{400 \hspace{0.1em} \mathrm{nm}}^{490 \hspace{0.1em} \mathrm{nm}} \frac{1}{\lambda^5} \frac{1}{\exp\left (\frac{hc}{\lambda k_B T}\right) - 1} d\lambda \hspace{3em} \mathrm{and} \hspace{3em} f_V = 2 h c^2 \int_{505 \hspace{0.1em} \mathrm{nm}}^{595 \hspace{0.1em} \mathrm{nm}} \frac{1}{\lambda^5} \frac{1}{\exp\left (\frac{hc}{\lambda k_B T}\right) - 1} d\lambda$$

Defining the dimensionless quantity

$$x = \frac{hc}{\lambda k_B T}$$

we have

$$dx = - \frac{hc}{\lambda^2 k_B T} d\lambda = - \frac{x^2k_BT}{hc} d\lambda$$

so that our integrals become

$$f_B = 2 h c^2 \int_{x(400 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(490 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{1}{(\frac{hc}{xk_BT})^5} \frac{1}{e^x - 1} \cdot -\frac{hc}{x^2k_BT} dx   \hspace{3em} \mathrm{and} \hspace{3em}f_B = 2 h c^2 \int_{x(505 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(595 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{1}{(\frac{hc}{xk_BT})^5} \frac{1}{e^x - 1} \cdot -\frac{hc}{x^2k_BT} dx $$

$$\Downarrow$$

$$f_B = -\frac{2(k_B T)^4}{h^3 c^2} \int_{x(400 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(490 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{x^3}{e^x - 1} dx \hspace{3em} \mathrm{and} \hspace{3em} f_V =  -\frac{2(k_B T)^4}{h^3 c^2} \int_{x(505 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(595 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{x^3}{e^x - 1} dx$$

$$\Downarrow$$

$$f_B = \frac{2(k_B T)^4}{h^3 c^2} \int_{x(490 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(400 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{x^3}{e^x - 1} dx \hspace{3em} \mathrm{and} \hspace{3em} f_V =  \frac{2(k_B T)^4}{h^3 c^2} \int_{x(595 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(505 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{x^3}{e^x - 1} dx$$

Time now for a word on the various constants/units. Fortunately, since all the color index cares about is the flux ratio, we needn't worry about the $\frac{2(k_B T)^4}{h^3 c^2}$ prefactors at all; in fact, let's make the definitions

$$I_B := \int_{x(490 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(400 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{x^3}{e^x - 1} dx \hspace{3em} \mathrm{and} \hspace{3em} I_V :=  \int_{x(595 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)}^{x(505 \hspace{0.1em} \mathrm{nm},\hspace{0.1em} T)} \frac{x^3}{e^x - 1} dx$$

so that

$$B - V = -2.5 \log_{10}
  \left(\frac{I_B}{I_V}\right)$$
  
(I'm not doing anything deep&mdash;I'm just trying to strip away "extraneous" information so that I'll ultimately have less to type in the program.)

As for the bounds of integration, the units we choose for the constants must respect the fact that the wavelengths are in nm. So we use $h = 4.135667696 \times 10^{-15} \hspace{0.1em} \mathrm{eV}\cdot\mathrm{s}$;
$c = 2.99 792 458 \times 10^{17} \hspace{0.1em} \mathrm{nm}/\mathrm{s}$; and $k_B = 8.617333262 \times 10^{-5} \hspace{0.1em} \mathrm{eV}/\mathrm{K}$. Let's just make sure everything works out:

$$[x] = \frac{[h][c]}{[\lambda][k_B][T]} = \frac{\mathrm{eV}\cdot\mathrm{s}\cdot\frac{\mathrm{nm}}{{\mathrm{s}}}}{\mathrm{nm} \cdot \frac{\mathrm{eV}}{\mathrm{K}} \cdot \mathrm{K}} = [1] \hspace{0.2em} \checkmark$$

One more thing: magnitudes have a zero-point (calibration); in particular, we will assume something Vega-like and define $B-V = 0$ for $T = 10000\hspace{0.1em}\mathrm{K}$. This means that we will shift our results by a constant offset.

```python
import numpy as np
import matplotlib.pyplot as plt

h = 4.135667696e-15 
c = 2.99792458e17
k_b = 8.617333252e-5

def x_from_lambda(lambda_bound_of_integration, T):
    return (h*c)/(lambda_bound_of_integration * k_b * T)

def integrand(x):
    return np.where(x == 0, 0, x**3 / (np.expm1(x)))

def b_minus_v_index(I_b, I_v):
    return -2.5 * np.log10(I_b/I_v)

def simpson(x):
    """composite Simpsons rule.  Here x is the vector
    of coordinate values we will evaluate the function
    at."""
    
    # the number of bins is one less than the number of
    # points
    N = len(x)-1
    
    # we require an even number of bins
    assert N % 2 == 0
    
    I = 0.0

    # loop over bins
    for n in range(0, N-1, 2):
        dx = x[n+1] - x[n]
        I += dx/3.0 * (integrand(x[n])+ 4 * integrand(x[n+1]) + integrand(x[n+2]))
    return I

sample_temperature_array = 10**np.linspace(np.log10(4000), np.log10(25000), 20) 
# 20 temperatures evenly spaced in log10(T) between 4000 K and 25000 K

list_of_b_minus_v_indices = []
nbins = 1000 # for how we settled on this number, see later commentary

calibration_temperature_vega = 10000
x_490_vega = x_from_lambda(490, calibration_temperature_vega)
x_400_vega = x_from_lambda(400, calibration_temperature_vega)
x_595_vega = x_from_lambda(595, calibration_temperature_vega)
x_505_vega = x_from_lambda(505, calibration_temperature_vega)
blue_xs_vega = np.linspace(x_490_vega, x_400_vega, nbins + 1)
visual_xs_vega = np.linspace(x_595_vega, x_505_vega, nbins + 1)
I_B_vega = simpson(blue_xs_vega)
I_V_vega = simpson(visual_xs_vega)
b_minus_v_index_vega = b_minus_v_index(I_B_vega, I_V_vega)


for given_temp in sample_temperature_array:
    x_490_given_temp = x_from_lambda(490, given_temp)
    x_400_given_temp = x_from_lambda(400, given_temp)
    x_595_given_temp = x_from_lambda(595, given_temp)
    x_505_given_temp = x_from_lambda(505, given_temp)
    blue_xs_given_temp = np.linspace(x_490_given_temp, x_400_given_temp, nbins + 1)
    visual_xs_given_temp = np.linspace(x_595_given_temp, x_505_given_temp, nbins + 1)
    I_B_given_temp = simpson(blue_xs_given_temp)
    I_V_given_temp = simpson(visual_xs_given_temp)
    b_minus_v_index_given_temp = b_minus_v_index(I_B_given_temp, I_V_given_temp)
    b_minus_v_index_given_temp -= b_minus_v_index_vega # constant offset adjustment 
    list_of_b_minus_v_indices.append(b_minus_v_index_given_temp)
```

```python
fig = plt.figure(dpi=1200) # extra argument for nice resolution
plt.rcParams['mathtext.fontset'] = 'stix' # commands to control fonts on graph
plt.rcParams['font.family'] = 'STIXGeneral'

ax = fig.add_subplot(111)
ax.scatter(sample_temperature_array, list_of_b_minus_v_indices, label = "Calculated values", marker = "*")
ax.legend()
ax.set_xlabel("Temperature ($K$)")
ax.set_ylabel("$B-V$")
ax.set_title("$B-V$ vs. $T$")

# now let's add some real data
real_temperatures = [30000, 9520, 7200, 6030, 5250, 3850]
real_indices = [-0.30, -0.02, 0.30, 0.58, 0.81, 1.40]
ax.scatter(real_temperatures, real_indices, label = "Real data (Carroll & Ostlie)", marker = "*")
ax.legend()
ax.grid()
```

How did we settle on $\texttt{nbins = 1000}$? Well, sort of arbitrarily, actually. In fact, even two bins would have worked just fine---observe how all the points in the graph below sit pretty much atop one another. The reason for this, I suspect, is that because we are integrating within such a narrow range (due to the filter application), the variation in the Planck function over the range of integration is so small that even relatively few bins capture the morphology of curve just fine. 

```python
def b_minus_v_indices_as_a_function_of_n_bins(n_bins):
    list_of_b_minus_v_indices = []
    for given_temp in sample_temperature_array:
        x_490_given_temp = x_from_lambda(490, given_temp)
        x_400_given_temp = x_from_lambda(400, given_temp)
        x_595_given_temp = x_from_lambda(595, given_temp)
        x_505_given_temp = x_from_lambda(505, given_temp)
        blue_xs_given_temp = np.linspace(x_490_given_temp, x_400_given_temp, n_bins + 1)
        visual_xs_given_temp = np.linspace(x_595_given_temp, x_505_given_temp, n_bins + 1)
        I_B_given_temp = simpson(blue_xs_given_temp)
        I_V_given_temp = simpson(visual_xs_given_temp)
        b_minus_v_index_given_temp = b_minus_v_index(I_B_given_temp, I_V_given_temp)
        b_minus_v_index_given_temp -= b_minus_v_index_vega # constant offset adjustment 
        list_of_b_minus_v_indices.append(b_minus_v_index_given_temp)
    return np.array(list_of_b_minus_v_indices)


list_of_possible_n_bins = [2**x for x in range(1, 11)]
# 2, 4, ... 1024

fig = plt.figure(dpi=1200)
ax = fig.add_subplot(111)
ax.legend() 
ax.set_xlabel("Temperature ($K$)")
ax.set_ylabel("$B-V$")
ax.set_title("$B-V$ vs. $T$ for Various Numbers of Bins in the Integration")
for possible_n_bins in list_of_possible_n_bins:
    b_minus_v_indices = b_minus_v_indices_as_a_function_of_n_bins(possible_n_bins)
    ax.scatter(sample_temperature_array, b_minus_v_indices, label = f"Results with {possible_n_bins} bins", s = 0.3)
    ax.legend() 
```
